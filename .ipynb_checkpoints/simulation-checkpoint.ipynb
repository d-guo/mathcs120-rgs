### Description
In part I, we simulate the random graphs

1. $G(n, \frac{1}{n^{1.1}})$
2. $G(n, \frac{1}{n})$
3. $G(n, \frac{4}{n})$
4. $G(n, \frac{\log(n)}{n})$

for various $n$. For each $n$, we run $m$ simulations and plot the number of triangles as a function of $n$, as well as the corresponding variance.

In part II, we simulate $G(n, \frac{\lambda}{n})$ with $\lambda = 2$ for various $n$ and plot the size of the largest component as a function of $n$.

### Imports

```python
import networkx as nx
import random
import matplotlib.pyplot as plt
from statistics import variance
from math import comb, log
```

## PART I

### Hyperparameters

```python
# set range of values for n [a, b)
a, b = 0, 1001

# set step size
s = 200

# set number of simulations for each n
m = 250
```

### main function for simulations

```python
# function for simulating a random graph with n nodes and probability p_n
# runs m simulations
def sim_single_RG(n, p_n, m):
    """
    inputs
    n: # of nodes
    p_n: a lambda to calculate probability of an edge as a function of n
    m: # of sims
    
    runs m simulations of the specified random graph
    
    returns tuple (average # of triangles, variance)
    """
    
    num_triangles = []
    
    for _ in range(m):
        if n in [0, 1, 2]:
            return 0, 0

        # initializes an empty graph
        G = nx.Graph()

        # create n nodes
        G.add_nodes_from(range(n))

        # create edges with probability p
        for i in range(n):
            for j in range(i + 1, n):
                if random.uniform(0, 1) < p_n(n):
                    G.add_edge(i, j)

        # count triangles
        # nx.algorithms.cluster.triangles(G1) returns a dictionary formatted as
        # {node_i: number of triangles with node_i as a vertex}
        # to find total triangles, we sum all of these and divide by 3
        num_triangles.append(int(sum(nx.algorithms.cluster.triangles(G).values()) / 3))
    
    return sum(num_triangles) / len(num_triangles), variance(num_triangles)
```

### wrapper function to call main function

```python
# function for simulating any type of graph for iterations in [a, b) with step size s
# type of graph determined by p_n lambda probability of edge wrt n
def run_sim(p_n):
    # run simulations for graph and format the data for plotting
    G_num_triangles = []
    G_variance = []

    for n in range(a, b, s):
        avg_num_triangles, var = sim_single_RG(n, p_n, m)
        G_num_triangles.append(avg_num_triangles)
        G_variance.append(var)
    
    return G_num_triangles, G_variance
```

## Simulation for Graph 1
$G(n, \frac{1}{n^{1.1}})$ 

```python
# graph is defined by p_n
p_n_1 = lambda n : 1 / (n ** 1.1)
G1_num_triangles, G1_variance = run_sim(p_n_1)
```

### Plot of Graph 1 Triangles

```python
plt.plot(range(a, b, s), G1_num_triangles)
plt.show()
```

### Plot of Expected Triangles of Graph 1

```python
plt.plot(range(a, b, s), [comb(n, 3) * (p_n_1(n) ** 3) if n not in [0, 1, 2] else 0 for n in range(a, b, s)])
plt.show()
```

### Plot of Graph 1 Variance

```python
plt.plot(range(a, b, s), G1_variance)
plt.show()
```

## PART II

### Hyperparameters

```python
# set range of values for n [a, b)
a, b = 0, 1001

# set step size
s = 200

# set number of simulations for each n
m = 250
```

### main function for simulations for largest connected component

```python
# function for simulating a random graph with n nodes and probability p_n
# runs m simulations
def sim_single_RG_c(n, p_n, m):
    """
    inputs
    n: # of nodes
    p_n: a lambda to calculate probability of an edge as a function of n
    m: # of sims
    
    runs m simulations of the specified random graph
    
    returns tuple (average size of largest connected component, variance)
    """
    
    largest_c_size = []
    
    for _ in range(m):
        if n in [0, 1, 2]:
            return 0, 0

        # initializes an empty graph
        G = nx.Graph()

        # create n nodes
        G.add_nodes_from(range(n))

        # create edges with probability p
        for i in range(n):
            for j in range(i + 1, n):
                if random.uniform(0, 1) < p_n(n):
                    G.add_edge(i, j)

        # count size of largest connected component
        # max(nx.algorithms.components.connected_components(G), key=len) returns the largest connected component
        largest_c_size.append(len(max(nx.algorithms.components.connected_components(G), key=len)))
    
    return sum(largest_c_size) / len(largest_c_size), variance(largest_c_size)
```

### wrapper function to call previous function

```python
# function for simulating any type of graph for iterations in [a, b) with step size s for the largest connected component
# type of graph determined by p_n lambda probability of edge wrt n
def run_sim_c(p_n):
    # run simulations for graph and format the data for plotting
    G_largest_c = []
    G_variance = []

    for n in range(a, b, s):
        largest_c_size, var = sim_single_RG_c(n, p_n, m)
        G_largest_c.append(largest_c_size)
        G_variance.append(var)
    
    return G_largest_c, G_variance
```

```python
# graph is defined by p_n
p_n = lambda n : 2 / n
G_largest_c, G_variance = run_sim_c(p_n)
```

### Plot of Graph Size of Largest Components

```python
plt.plot(range(a, b, s), G_largest_c)
plt.show()
```

### Plot of Expected Size of Largest Components of Graph

```python
plt.plot(range(a, b, s), [0.797 * n for n in range(a, b, s)])
plt.show()
```

### Plot of Graph Variance

```python
plt.plot(range(a, b, s), G_variance)
plt.show()
```
